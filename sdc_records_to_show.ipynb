```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv


path = "/content/sample_data/energy.csv"
electricity=pd.read_csv(path)


print(electricity)

print(electricity.shape)
print(electricity.head(5))
print(electricity.tail(4))
print(electricity.info())
print(electricity.describe())


print("column Data")
print(electricity.columns)

print(electricity[electricity['State'].str.startswith('M')])
print(electricity[electricity['State'].str.endswith('a')])
rcb=electricity[electricity['State']=='Manipur']
print(rcb)


print(electricity.sort_values(by='State'))


print(np.max(electricity['Consumption']))
print(np.min(electricity['Consumption']))
print(np.mean(electricity['Consumption']))
print(np.median(electricity['Consumption']))
print(np.std(electricity['Consumption']))


# Assuming you want to plot data for 'Andhra' as state1_data
print(electricity[electricity['State']=='Andhra']['Consumption'].max()) # Use () for calling max function
print(electricity[electricity['State']=='Andhra']['Consumption'].min()) # Use () for calling min function

print("Andhra Data")

state1_data=electricity[electricity['State']=='Andhra'] # Define state1_data
state2_data=electricity[electricity['State']=='Karnataka']

state4_data=electricity[electricity['State']=='Telengana']

state5_data=electricity[electricity['State']=='Tirpura']

fig,axes = plt.subplots(2)
axes[0].plot(state1_data['Data'],state1_data['Consumption'],marker='*',color='red') # Change market to marker
axes[1].plot(state2_data['Data'],state2_data['Consumption'],marker='^',color='blue') # Change market to marker
```
